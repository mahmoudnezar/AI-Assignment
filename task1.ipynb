```python
# Constants
G = 6.674 * 10**-11

# Inputs
M = float(input("Enter the mass of the first object: "))
m = float(input("Enter the mass of the second object: "))
r = float(input("Enter the distance between the two objects: "))

# Calculation
force = (G * M * m) / (r**2)

# Output
print("The gravitational force between the two objects is:", force)
```
